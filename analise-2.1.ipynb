Instalando as bibliotecas a serem utilizadas

```r
install.packages("readxl")
install.packages("ggplot2")


library(readxl)
library(ggplot2)
```

Lendo o arquivo com os dados

```r
caminho_arquivo <- "prioridades-projetos-mestrado.xlsx"

dados <- read_excel(caminho_arquivo)
dados <- as.data.frame(dados)

head(dados)
```

Verificando se exitem strings com 'NA' e NA. Fazendo o tratamento

```r
dados[dados == 'NA']
```

```r
dados[dados == NA]
```

```r
dados_tratados <- na.omit(dados)
dados_tratados
```

Utilizando o metodo de colocar valores dependendo da ordem de acordo aao caminho das pedras

```r
# labels do likert
likert = c("Alta prioridade","Alta-média prioridade", "Média prioridade", "Média-baixa prioridade", "Baixa prioridade")
likert
```

```r
dados_numericos <- dados_tratados
```

```r
dados_numericos[] = lapply(dados_tratados, function(x) as.numeric(factor(x, levels = likert, ordered = TRUE)))
dados_numericos
```

Utilização do table para a contagem de frequências


Usando colnames para criar um vetor com o nome das colunas

```r
table(dados_numericos$Doc1)
```

```r
nome_colunas <- colnames(dados_numericos)
nome_colunas
```

```r
lista_contagens <- lapply(dados_numericos[nome_colunas], table)
lista_contagens
```

comparando valor de string. Uma vez que o resultado de table é em string

```r
somar_prioridade_alta <- function(tabela) {
  valor_1 <- tabela["1"]
  valor_2 <- tabela["2"]

  valor_1[is.na(valor_1)] <- 0
  valor_2[is.na(valor_2)] <- 0

  return(unname(valor_1 + valor_2))
}
```

```r
somas_finais_colunas <- sapply(lista_contagens, somar_prioridade_alta)
somas_finais_colunas
```

Calculando o valor maximo e o nome da sub area desse valor maximo

```r
max_valor <- max(somas_finais_colunas)
sub_area_max <- names(somas_finais_colunas)[which.max(somas_finais_colunas)]
sub_area_max
max_valor
```

Transformando os vetores em um data.frame para a plotagem do gráfico

```r
df_grafico <- data.frame(
  Sub_Area = names(somas_finais_colunas),
  Soma_Alta_Media = as.numeric(somas_finais_colunas)
)
df_grafico
```

Plotando o gráfico com a largura ajustado. No goole Colab fica pequeno

```r

options(repr.plot.width = 18, repr.plot.height = 6)

ggplot(data = df_grafico, aes(x = Sub_Area, y = Soma_Alta_Media)) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = Soma_Alta_Media), vjust = -0.5, size = 5) +
  geom_point(size = 0) +
  labs(title = "Comparação de Prioridades Altas e altras médias",
       x = "Sub-Área",
       y = "Total de prioridades") +
  theme_minimal()
```

Fazendo a ordenação do gráfico para vizualizar as sub áres de maior prioridade a menor prioridade

Ao observar o gráfico com todas as somatorias das colunas:


*   A maioria das somatorias se concentra entre 30 e 34
*   Com um pico no valor 39 na terceira sub-área, com isso se torna o com mais prioridade entre todas
*   Existem 2 sub-áreas que estao com valor de 21 ou sejam as que devem ser menor priorizadas.
*  A priorização se encontra bem dispersa em valores



```r

options(repr.plot.width = 18, repr.plot.height = 6)

ggplot(data = df_grafico, aes(
    x = reorder(Sub_Area, -Soma_Alta_Media),
    y = Soma_Alta_Media
  )) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = Soma_Alta_Media), vjust = -0.5, size = 5) +
  labs(
    title = "Comparação de Prioridades Altas e Altas-médias (Ordenado)",
    x = "Sub-Área",
    y = "Total de prioridades"
  ) +
  theme_minimal()
```

Ao observar o gráfico ordenado com todas as somatorias das colunas:


*   A maioria das somatorias estão agrupadas em grupos de 4 ou 5. Na qual as exeções são o maior valor que é 39 e o menor valor 21, além dos valores 36, 35, 25 e 24
* Existem muitas sub-áres com a prioridade alta
